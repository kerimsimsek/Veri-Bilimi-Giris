```python
import pandas as pd
import numpy as np
```

```python
#değişken ismi = numpy.rastgele sayı.tam sayı(rastgele oluşacak sayı kaça kadar olsun,kaç tane veri olsun)
a = np.random.randint(10,size=6)
b = np.random.randint(10,size=6)
c = np.random.randint(10,size=6)
d = np.random.randint(10,size=6)

#değişken ismi= pandas.veri çerçevesi({"sütun adı":veri(bir önceki satırda oluşturduğum veriyi çağırıyorum)})
data = pd.DataFrame({"s1":a,"s2":b,"s3":c,"s4":d})
```

```python
data
```

```python
data2 = data + 10
```

```python
data2
```

```python
pd.concat([data,data2]) #satıra ekledi fakat devamı olacak şekilde indekslemedi.
```

```python
pd.concat([data,data2],ignore_index=True) #ignore_index komutu sanki devamıymış gibi ekledi.
```

```python
pd.concat([data,data2],axis=1) #sütun başlıkları tekrar etti.
```

```python
pd.concat([data,data2],axis=1, ignore_index=True) #indeksleri yoksaydı yani sütun ve satır isimlerini sildi.
```
