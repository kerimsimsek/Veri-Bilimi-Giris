```python
import pandas as pd
import numpy as np
```

```python
#değişken ismi = numpy.rastgele sayı.tam sayı(rastgele oluşacak sayı kaça kadar olsun,kaç tane veri olsun)
a = np.random.randint(10,size=6)
b = np.random.randint(10,size=6)
c = np.random.randint(10,size=6)
d = np.random.randint(10,size=6)
```

```python
#değişken ismi= pandas.veri çerçevesi({"sütun adı":veri(bir önceki satırda oluşturduğum veriyi çağırıyorum)})
data = pd.DataFrame({"s1":a,"s2":b,"s3":c,"s4":d})
```

```python
data
```

```python
data.drop(2,axis=0) #datanın 2.satırını sildim.(axis=0 satırı ifade etmek demek)
```

```python
data #datadan 2.satırı silmeme rağmen datayı çağırınca geri geliyor bunu kalıcı hale getirmek için inplace komutu kullanılır.
```

```python
data.drop(2,axis=0, inplace=True) #inplace ile kalıcı olarak satırı sildim.
```

```python
data
```

```python
data.drop("s1",axis=1) #datanın s1 başlıklı sütununu sildim. (axis=1 sütunu ifade etmek için)
```

```python
data #datanın sütununu silmeme rağmen geri geliyor çünkü kalıcı yapmadım.
```

```python
data.drop("s1",axis=1, inplace=True) #datanın sütununu inplace komutu ile kalıcı olarak sildim.
```

```python
data
```

```python
data.drop([3,4],axis=0) #birden çok satırı silmek istediğimde köşeli parantez içinde indeksini yazıyorum.
```

```python
data.drop(["s2","s3"],axis=1) #birden çok sütunu silmek için köşeli parantez içinde sütun başlıkları yaz.
```
