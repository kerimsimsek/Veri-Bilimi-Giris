```python
import pandas as pd
import numpy as np
```

```python
#değişken ismi = numpy.rastgele sayı.tam sayı(rastgele oluşacak sayı kaça kadar olsun,kaç tane veri olsun)
a = np.random.randint(10,size=6)
b = np.random.randint(10,size=6)
c = np.random.randint(10,size=6)
d = np.random.randint(10,size=6)

#değişken ismi= pandas.veri çerçevesi({"sütun adı":veri(bir önceki satırda oluşturduğum veriyi çağırıyorum)})
data = pd.DataFrame({"s1":a,"s2":b,"s3":c,"s4":d})
```

```python
data
```

```python
data>5 #verilerin 5'den büyük olanları true, küçük ve eşit olanları false değerini döndürür.
```

```python
data[data>5] #verilerin 5'den büyük olanları yazar, küçük ve eşit olanları NaN değerini döndürür.
```

```python
data[data.s2>5] #2.sütundan sadece 5'den büyük olan verileri getir.
```

```python
data[data.s2>5]["s2"] #2.sütundan sadece 5'den büyük olan verileri getirir ve sadece s2 tablosunu gösterir.
```
